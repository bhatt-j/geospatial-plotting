```python
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# nybb - map of New York boroughs
ny = gpd.read_file(gpd.datasets.get_path('nybb')) 
ny.head()
```

```python
ny.plot()
```

```python
ny.crs
```

```python
# Web Mercator

ny_wm = ny.to_crs(epsg = 3857)
```

```python
import contextily as cx
```

```python
ax_ny = ny_wm.plot(figsize=(10,10), alpha=0.5, edgecolor = "k")
cx.add_basemap(ax_ny)
```

By default, contextily uses the Stamen Terrain style. 
* Different style using - <b>cx.providers</b>:

```python
fig = plt.figure(figsize=(16,9))
axx_ny = plt.subplot()
ny.plot(color='none',edgecolor='green', ax = axx_ny)
cx.add_basemap(ax = ax_ny, source=cx.providers.Stamen.Toner)
```
